```python
#TMDB provides a json file that lists out all movie id's. 900803 is the highest one

import json
import requests
from pandas import json_normalize


file = 'movie_ids_11_21_2021.json'

jsondf = pd.read_json('movie_ids_11_21_2021.json',lines = True)
jsondf2 = jsondf.sort_values('id',ascending=True)
jsondf3 = jsondf2[jsondf2['popularity'] > 4]
idlist = jsondf3['id'].tolist()
len(idlist)
idlist_sample = idlist[:5]
```

```python
headers = ['id','title','overview','original_language','budget','revenue','release_date',
           'vote_average','vote_count','director','lead_actor_1','lead_actor_2','lead_actor_3',
           'lead_actor_4'
          ]
```

```python
def inlist(j,i):
    try:
        return tmdb.Movies(j).info()[i]
    except:
        pass
```

```python
#Step 1: Pandas DataFrame of movie information
headers1 = ['id','title','overview','original_language','budget','revenue','release_date',
           'vote_average','vote_count']

movie_info = [[inlist(j,i) for i in headers1] for j in idlist]

df1 = pd.DataFrame(movie_info, columns = headers1)

df1.dropna(how='all',inplace=True)
df1.to_csv(f'data/movie_info.csv',index=False)
df1
```

```python
def inlistdir(l):
    try:
        return [k for k in tmdb.Movies(l).credits()['crew'] if k['job'] == 'Director'][0]['name']
    except:
        pass
```

```python
director = [inlistdir(l) for l in idlist]
```

```python
df2 = pd.DataFrame({'id': idlist, 'director': director})
df2.dropna(how='all',inplace=True)
df2.to_csv(f'data/director.csv',index=False)
df2.head()
```

```python
def inlistactor(m):
    try:
        return [tmdb.Movies(m).credits()['cast'][n]['name'] for n in range(0,4)]
    except:
        pass
```

```python
actors = [inlistactor(m) for m in idlist]
```

```python
actors2 = actors.copy()
actors2 = [['x','x','x','x'] if v is None else v for v in actors2]
actors2[50:55]
```

```python
headers3 = ['lead_actor_1','lead_actor_2','lead_actor_3','lead_actor_4','id_init']
df3 = pd.DataFrame(actors2,columns = headers3)
df3['id'] = idlist
df3 = df3[['id','lead_actor_1','lead_actor_2','lead_actor_3','lead_actor_4']]
df3.to_csv(f'data/actors.csv',index=False)
df3.head()
```

```python
df_merge = pd.merge(df1, df2,how='inner',left_on='id',right_on='id')
df_merge = pd.merge(df_merge,df3,how='inner',left_on='id',right_on='id')
df_merge.to_csv(f'data/tmdb_data.csv',index=False)
df_merge.head()
```
